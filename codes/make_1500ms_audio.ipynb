```python
from pydub import AudioSegment
import os
from shutil import copyfile
```

```python
def add_silence(sound):
    n = 1500
    audio_length = len(sound)
    extra = audio_length%n
    silence_duration = n - extra
    silence_part = AudioSegment.silent(duration = silence_duration)
    new_sound = sound + silence_part
    return new_sound
    
```

```python


clip_folder = '/home/user/Sahil/donatecry_clips_5sec_28_02_2021/tired'
max = 0
min = 1000000
i = 0
for filename in os.listdir(clip_folder):
    i = i+1
    print(filename)
    fname = clip_folder + '/' + filename
    sound = AudioSegment.from_wav(fname)
    audio_length = len(sound)
    new_sound = add_silence(sound)
    #print(len(new_sound))
    export_path = '/home/user/Sahil/donatecry_1500_data/tired' +'/'+filename
    new_sound.export(export_path,format = 'wav')
    
    #print(audio_length)
    
    if(max<len(new_sound)):
        max = audio_length
    if(min>len(new_sound)):
        min = audio_length
    #print(max)
    #print(i)
print(max)
print(min)
    
```

```python
def slicing_big_chunks(sound,folder,filename):
    n = 1500
    audio_length = len(sound)
    i = 0
    j = 0
    while(i<audio_length):
        if(audio_length > n):
            j = j + 1
            newfilename = filename + '_' + str(j) 
            print(newfilename)
            path = folder + '/' + newfilename + '.wav'
            new_sound = sound[i:(i+n-1)]
            new_sound.export(path, format="wav")
            i = i+n
        else:
            i = audio_length
            print("cannot slice")
```

```python
clip_folder = "/home/user/Sahil/donatecry_1500_data/tired"
target_folder = "/home/user/Sahil/donatecry_1500_data/final/tired"

for filename in os.listdir(clip_folder):
    fname = clip_folder + '/' + filename
    sound = AudioSegment.from_wav(fname)
    noext_name = os.path.splitext(filename)[0]
    slicing_big_chunks(sound,target_folder,noext_name)
```
