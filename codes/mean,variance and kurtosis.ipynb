```python
import numpy as np
import pandas as pd
import librosa
from scipy.stats import kurtosis,skew
```

change to add skew

```python
def get_features(dct_file,target_path):
    x =np.load(dct_file)
    #print(x.shape)
    mean_arr = np.mean(x,axis = -1)
    #print(y.shape)
    var_arr = np.var(x,axis = -1)
    kurt_arr = kurtosis(x,axis = -1)
    skew_arr = skew(x,axis = -1)
    feat = np.concatenate((mean_arr,var_arr,kurt_arr,skew_arr))
    print(feat.shape)
    np.save(target_path,feat)
```

```python
import os

clip_folder1 = r'/home/user/Sahil/orig_features_68'

for filename in os.listdir(clip_folder1):
    fname = clip_folder1 + '/' + filename
    target_file = r'/home/user/Sahil/orig_mean_var_kurt_skew' +'/' + filename[0:len(filename)-4]
    get_features(fname,target_file)
    print(filename)
```
