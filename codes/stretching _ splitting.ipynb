```
import librosa
import numpy as np
import matplotlib.pyplot as plt
import os

class AudioAugmentation:
    def read_audio_file(self, file_path):
        data = librosa.core.load(file_path,sr=None)[0] #, sr=16000
        return data

    def write_audio_file(self, file, data, sample_rate=8000):
        librosa.output.write_wav(file, data, 8000,norm=False)

    def stretch(self, data, rate=0.25):
        #input_length = 110250
        data = librosa.effects.time_stretch(data, rate)
        # if len(data) > input_length:
        #     data = data[:input_length]
        # else:
        #     data = np.pad(data, (0, max(0, input_length - len(data))), "constant")
        return data

aa = AudioAugmentation()

directory1 = '/content/drive/My Drive/clips/belly_pain'
for filename in os.listdir(directory1):
  if filename.endswith(".wav") :
    
   fname = directory1 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   data_stretch = aa.stretch(data,0.25)
   aa.write_audio_file('/content/drive/My Drive/clips/belly_pain/stretched_20s/%s.wav' % newfname, data_stretch)


directory3 = '/content/drive/My Drive/clips/discomfort'
for filename in os.listdir(directory3):
  if filename.endswith(".wav") :
    
   fname = directory3 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   data_stretch = aa.stretch(data,0.25)
   aa.write_audio_file('/content/drive/My Drive/clips/discomfort/stretched_20s/%s.wav' % newfname, data_stretch)



directory4 = '/content/drive/My Drive/clips/tired'
for filename in os.listdir(directory4):
  if filename.endswith(".wav") :
    
   fname = directory4 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   data_stretch = aa.stretch(data,0.25)
   aa.write_audio_file('/content/drive/My Drive/clips/tired/stretched_20s/%s.wav' % newfname, data_stretch)

directory2 = '/content/drive/My Drive/clips/burping'
for filename in os.listdir(directory2):
  if filename.endswith(".wav") :
    
   fname = directory2 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   data_stretch = aa.stretch(data,0.25)
   aa.write_audio_file('/content/drive/My Drive/clips/burping/stretched_20s/%s.wav' % newfname, data_stretch)
```

```
from pydub import AudioSegment

directory2 = '/content/drive/My Drive/clips/burping/stretched_20s'
for filename in os.listdir(directory2):
  if filename.endswith(".wav") :
    fname = directory2 + '/' + filename
    string = '.wav'
    newfname = filename[:-(len(string))]

    sound = AudioSegment.from_file(fname, format="wav")

    # split sound in 5-second slices and export
    for i, chunk in enumerate(sound[::5000]):
      with open("/content/drive/My Drive/clips/burping/20split/%s%d.wav" % (newfname , (i+3)), "wb") as f:
         chunk.export(f, format="wav")

directory1 = '/content/drive/My Drive/clips/belly_pain/stretched_20s'
for filename in os.listdir(directory1):
  if filename.endswith(".wav") :
    fname = directory1 + '/' + filename
    string = '.wav'
    newfname = filename[:-(len(string))]

    sound = AudioSegment.from_file(fname, format="wav")

    # split sound in 5-second slices and export
    for i, chunk in enumerate(sound[::5000]):
      with open("/content/drive/My Drive/clips/belly_pain/20split/%s%d.wav" % (newfname , (i+3)), "wb") as f:
         chunk.export(f, format="wav")         

directory3 = '/content/drive/My Drive/clips/discomfort/stretched_20s'
for filename in os.listdir(directory3):
  if filename.endswith(".wav") :
    fname = directory3 + '/' + filename
    string = '.wav'
    newfname = filename[:-(len(string))]

    sound = AudioSegment.from_file(fname, format="wav")

    # split sound in 5-second slices and export
    for i, chunk in enumerate(sound[::5000]):
      with open("/content/drive/My Drive/clips/discomfort/20split/%s%d.wav" % (newfname , (i+3)), "wb") as f:
         chunk.export(f, format="wav")

directory4 = '/content/drive/My Drive/clips/tired/stretched_20s'
for filename in os.listdir(directory4):
  if filename.endswith(".wav") :
    fname = directory4 + '/' + filename
    string = '.wav'
    newfname = filename[:-(len(string))]

    sound = AudioSegment.from_file(fname, format="wav")

    # split sound in 5-second slices and export
    for i, chunk in enumerate(sound[::5000]):
      with open("/content/drive/My Drive/clips/tired/20split/%s%d.wav" % (newfname , (i+3)), "wb") as f:
         chunk.export(f, format="wav")         
```
