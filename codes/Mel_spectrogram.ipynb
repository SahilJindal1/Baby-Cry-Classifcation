```python
import librosa.display

import numpy as np
import pandas as pd
import librosa
```

```python
def graph_spectrogram(wav_file,target_path):
    y,rate = librosa.load(wav_file)
    print(rate)
    window_size = 1024
    window = np.hanning(window_size)
    spec = librosa.feature.melspectrogram(y=y, sr=44100, S=None, n_fft=window_size, hop_length=512, window=window, center=True, pad_mode='reflect', power=2.0)
    np.save(target_path,spec)
    
```

```python

import os

clip_folder1 = r'/home/user/Sahil/final_1500ms_donatecry/tired'

for filename in os.listdir(clip_folder1):
    fname = clip_folder1 + '/' + filename
    target_file = r'/home/user/Sahil/donatecry_mel/tired' +'/' + filename[0:len(filename)-4]
    graph_spectrogram(fname,target_file)
    print(filename)
```
