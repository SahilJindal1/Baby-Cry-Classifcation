```
import librosa
import numpy as np
import matplotlib.pyplot as plt
import os

class AudioAugmentation:
    def read_audio_file(self, file_path):
        data = librosa.core.load(file_path,sr=None)[0] #, sr=16000
        return data

    def write_audio_file(self, file, data, sample_rate=8000):
        librosa.output.write_wav(file, data, 8000,norm=False)

    def plot_time_series(self, data):
        fig = plt.figure(figsize=(14, 5))
        plt.title('Raw wave ')
        plt.ylabel('Amplitude')
        plt.plot(np.linspace(0, 5, len(data)), data)
        plt.show()

    def add_noise(self, data):
        noise = np.random.randn(len(data))
        data_noise = data + 0.005 * noise
        return data_noise

    def shift(self, data):
        return np.roll(data, 8000)

    # def stretch(self, data, rate=1.1):
    #     #input_length = 110250
    #     data = librosa.effects.time_stretch(data, rate)
    #     # if len(data) > input_length:
    #     #     data = data[:input_length]
    #     # else:
    #     #     data = np.pad(data, (0, max(0, input_length - len(data))), "constant")
    #     return data

# Create a new instance from AudioAugmentation class
aa = AudioAugmentation()

# Read and show cat sound


directory1 = '/content/drive/My Drive/clips/burping'
for filename in os.listdir(directory1):
  if filename.endswith(".wav") :
    
   fname = directory1 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   #aa.plot_time_series(data)
   # Adding noise to sound
   data_noise = aa.add_noise(data)
   #aa.plot_time_series(data_noise)
   # Shifting the sound
   data_roll = aa.shift(data)
   #aa.plot_time_series(data_roll)
  #  # Stretching the sound
  #  data_stretch = aa.stretch(data, 0.8)
  #  #aa.plot_time_series(data_stretch)
   # Write generated cat sounds
   aa.write_audio_file('/content/drive/My Drive/clips/burping/augmented/%s1.wav' % newfname, data_noise)
   aa.write_audio_file('/content/drive/My Drive/clips/burping/augmented/%s2.wav' % newfname, data_roll)
  #  aa.write_audio_file('/content/drive/My Drive/clips/burping/augmented/%s3.wav' % newfname, data_stretch)

#similarly for other directories
directory2 = '/content/drive/My Drive/clips/belly_pain'
for filename in os.listdir(directory2):
  if filename.endswith(".wav") :
   fname = directory2 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   #aa.plot_time_series(data)
   # Adding noise to sound
   data_noise = aa.add_noise(data)
   #aa.plot_time_series(data_noise)
   # Shifting the sound
   data_roll = aa.shift(data)
   #aa.plot_time_series(data_roll)
  #  # Stretching the sound
  #  data_stretch = aa.stretch(data, 0.8)
  #  #aa.plot_time_series(data_stretch)
   # Write generated cat sounds
   aa.write_audio_file('/content/drive/My Drive/clips/belly_pain/augmented/%s1.wav' % newfname, data_noise)
   aa.write_audio_file('/content/drive/My Drive/clips/belly_pain/augmented/%s2.wav' % newfname, data_roll)
  #  aa.write_audio_file('/content/drive/My Drive/clips/belly_pain/augmented/%s3.wav' % newfname, data_stretch)

directory3 = '/content/drive/My Drive/clips/tired'
for filename in os.listdir(directory3):
  if filename.endswith(".wav") :
   fname = directory3 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   #aa.plot_time_series(data)
   # Adding noise to sound
   data_noise = aa.add_noise(data)
   #aa.plot_time_series(data_noise)
   # Shifting the sound
   data_roll = aa.shift(data)
   #aa.plot_time_series(data_roll)
   # Stretching the sound
  #  data_stretch = aa.stretch(data, 0.8)
   #aa.plot_time_series(data_stretch)
   # Write generated cat sounds
   aa.write_audio_file('/content/drive/My Drive/clips/tired/augmented/%s1.wav' % newfname, data_noise)
   aa.write_audio_file('/content/drive/My Drive/clips/tired/augmented/%s2.wav' % newfname, data_roll)
  #  aa.write_audio_file('/content/drive/My Drive/clips/tired/augmented/%s3.wav' % newfname, data_stretch)

directory4 = '/content/drive/My Drive/clips/discomfort'
for filename in os.listdir(directory4):
  if filename.endswith(".wav") :
   fname = directory4 + '/' + filename
   string = '.wav'
   newfname = filename[:-(len(string))]
   data = aa.read_audio_file(fname)
   #aa.plot_time_series(data)
   # Adding noise to sound
   data_noise = aa.add_noise(data)
   #aa.plot_time_series(data_noise)
   # Shifting the sound
   data_roll = aa.shift(data)
   #aa.plot_time_series(data_roll)
   # Stretching the sound
   #data_stretch = aa.stretch(data, 0.8)
   #aa.plot_time_series(data_stretch)
   # Write generated cat sounds
   aa.write_audio_file('/content/drive/My Drive/clips/discomfort/augmented/%s1.wav' % newfname, data_noise)
   aa.write_audio_file('/content/drive/My Drive/clips/discomfort/augmented/%s2.wav' % newfname, data_roll)
  #  aa.write_audio_file('/content/drive/My Drive/clips/discomfort/augmented/%s3.wav' % newfname, data_stretch)

```
